```python
import math
import os
import random
import numpy as np
import matplotlib
import glob
# # saving high quality figures
# matplotlib.use('agg')
# from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from numpy import genfromtxt
import pandas as pd

np.seterr(divide='ignore', invalid='ignore')
```

# Use csv files in the Results\\**games**\results\rews_and_steps\rewards_per_episode*.csv

```python
if __name__ == '__main__':
    files = [ 'rews/rewards_per_episode500.csv'
]
    all_data = np.empty(0)
    for file in files:
        a_data = genfromtxt(file, delimiter=',')
        all_data = np.append(all_data, a_data)
all_data.shape
```

```python
##
%matplotlib inline
x = pd.DataFrame(all_data)
y = x.rolling(window=10).mean()

plt.figure(figsize=(10,8))
plt.plot(range(len(y)), y)
plt.title('average reward vs eposide. Data is smoothed over 10 eposide')
plt.xlabel('eposide')
plt.ylabel('Average eposide reward')
```
